```python
import pandas as pd
dataset=  pd.read_csv('C:/Users/Gaelim/Desktop/hr_data.csv')
dataset.head()
```

```python
#Load in my dependencies  
import pandas as pd
import numpy as np 
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
#change the variable for the dataset to save time
df = dataset
# convert the categorial variable into to numerical varaibles and set target and features
df = pd.get_dummies(df,drop_first =True)
X= df.drop('Quit the Company',axis=1) 
y =df['Quit the Company'] 
#split the data 
X_train,X_test,y_train,y_test = train_test_split(X,y)
#initialize the classifier and fit the data to it
clf= DecisionTreeClassifier()
# train the model 
clf.fit(X_train, y_train)
#create a prediction 
y_pred = clf.predict(X_test)
#visualize the data 
class_names=['Stayed','Quit']
plt.figure(figsize=(12,12))
tree.plot_tree(clf,class_names=class_names,feature_names=X.columns, filled=True,fontsize=6)
plt.show()
```
